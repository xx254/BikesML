# Calculate the hourly change of number of bikes (diff)

```python
import pandas as pd
```

```python
df = pd.read_csv("data/trip_data.csv")
df.head()
```

```python
df["hourly"] = [ i[:13] for i in df["Start Date"]]
df["daily"] = [ i[:11] for i in df["Start Date"]]
df.head()
```

```python
df1=df.groupby(['End Station', 'hourly']).size().reset_index(name='count')
df1.columns=['Station','hourly','count']
df2=df.groupby(['Start Station', 'hourly']).size().reset_index(name='count')
df2.columns=['Station','hourly','count']
s1 = pd.merge(df1, df2, how='left', on=['Station', 'hourly'])
s1 
```

```python
s1 = s1.fillna(0)
s1['diff'] = s1['count_y'] - s1['count_x']
s1['month'] = [ i[:2] for i in s1["hourly"]]
s1['day'] = [ i[3:5] for i in s1["hourly"]]
s1['year'] = [ i[6:10] for i in s1["hourly"]]
s1['hour'] = [ i[11:] for i in s1["hourly"]]
s1['Date'] = s1['day'] + '/' + s1['month'] + '/' + s1['year']
s1.head(10)
```

```python
w = pd.read_csv("data/weather_data.csv")
w.head()
```

```python
s2 = pd.merge(s1, w, how='left', on=['Date'])
s2 = s2.drop(['Events', 'count_x', 'count_y', 'day', 'month', 'year', 'Date', 'hourly','Zip'], axis=1)
s2 = s2.dropna(axis=0, how='any')
s2.head()
```

```python
s2.dtypes
```

# Use Machine Learning Algorithm (Random Forest) to see the most significant factor to influence the change of number(diff)

```python
# exclude "hour"
feature1 = s2.columns[0:1] | s2.columns[3:20]
feature1
# inclue "hour"
feature2 = s2.columns[0:1] | s2.columns[2:20]
feature2
```

```python
# train and test
len = s2['Station'].count()
train = s2[0: int(len * 0.8)]
test = s2[int(len * 0.8):]
```

```python
y = pd.factorize(train['diff'])[0]
y
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.datasets import load_iris
from sklearn.metrics import mean_squared_error

iris = load_iris()

clf = RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[feature1], y)
```

```python
list(zip(train[feature1], clf.feature_importances_))
```

### Discovery: 'Max VisibilityMiles' and 'Station' significantly influence the number of bikes

```python
pred = clf.predict(test[feature1])
```

```python
clf = RandomForestClassifier(n_jobs=2, random_state=0)
clf.fit(train[feature2], y)
list(zip(train[feature2], clf.feature_importances_))
```

### Discovery: the influence of 'Max VisibilityMiles' significantly reduced and 
###           'Station' and 'Hour' are the two biggest influencer of the number of bikes

```python
from sklearn.metrics import mean_squared_error
```

```python
y_test = pd.factorize(test['diff'])[0]
predictions = clf.predict(test[feature2])
mse = mean_squared_error(y_test,predictions)
test['predictions']=predictions
print("MSE using RandomForestClassifier:    ")
print(mse)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
train = s2.sample(frac=.8)
test = s2.iloc[~s2.index.isin(train.index)]
features = list(s2.columns)
features.remove('diff')
model = LinearRegression()
model.fit(train[features],train['diff'])
predictions = model.predict(test[features])
mse = mean_squared_error(test['diff'],predictions)
print("MSE using LinearRegression:    ")
print(mse,'\n')
```

# Decision Tree Regression 

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(min_samples_leaf=5)
model.fit(train[features],train['diff'])
predictions=model.predict(test[features])
mse=mean_squared_error(test['diff'],predictions)
print("MSE using DecisionTreeRegression:    ")
print(mse,'\n')
```
